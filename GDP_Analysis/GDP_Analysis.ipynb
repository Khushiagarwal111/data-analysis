```python
final_data =[]


for country_name in df['Country Name'].unique():
    cn=df[df['Country Name']== country_name]

    data=cn.values
    gdp_change=[0]

    for i in range(1,len(data)):
        prev=data[i-1][3]
        cur=data[i][3]

        gdp_change.append(round(((cur-prev)/prev)*100,2))
    cn=cn.assign(GDP=gdp_change)
    final_data.append(cn)
    # print(counrty)
```

```python
new_df=pd.concat(final_data, axis=0)
new_df
```

```python
df = pd.concat(final_data, axis = 0)
```

```python
def compare_gdp(txt,):
  dfs=[]
  for i in txt:
    dfs.append(df[df['Country Code']==i])
    df_four_countries=pd.concat(dfs, axis=0)
    fig_compare=px.line(df_four_countries, x='Year', y='Value', title=" GDP Comparison - " + '|'.join(txt), color='Country Code')

    pyo.plot(fig_compare, filename= i+ ".html")
```
